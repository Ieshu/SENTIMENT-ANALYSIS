```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter import messagebox,simpledialog
import sqlite3
from datetime import datetime
from PIL import ImageTk,Image #pip install pillow
from tkinter import filedialog
import os
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd

# REMOVING STOPWORDS

words=stopwords.words("english")
words.remove('not')
words.remove("don't")
words.remove("didn't")
words.remove("wasn't")
words.remove("haven't")
words.remove("hasn't")
words.remove("hadn't")
words.remove("weren't")
words.remove("wouldn't")
words.remove("couldn't")
words.remove("shouldn't")
ps=PorterStemmer()

win=Tk()
win.resizable(width=False,height=True)
win.state('zoomed')
bgclr='#00ff80'
win.configure(bg=bgclr)
title=Label(win,text='Restaurant Review Analysis',font=('',50,'bold'),fg='black',bg=bgclr)
title.pack()

# CREATE HOMESCREEN

def homescreen(prvfrm=None):
    if(prvfrm!=None):
        prvfrm.destroy()
    frm=Frame(win,bg='yellow')
    frm.place(x=0,y=100,relwidth=1,relheight=1)
    user_lbl=Label(frm,text='Username:',font=('',20,''),fg='blue',bg='red')
    user_lbl.place(x=500,y=100)
    
    pass_lbl=Label(frm,text='Password:',font=('',20,''),fg='blue',bg='red')
    pass_lbl.place(x=500,y=150)
    
    user_entry=Entry(frm,bg='powder blue',bd=5,width=15,font=('',15,''))
    user_entry.place(x=680,y=100)
    user_entry.focus()
    
    pass_entry=Entry(frm,show='*',bg='powder blue',bd=5,width=15,font=('',15,''))
    pass_entry.place(x=680,y=150)
    
    login_btn=Button(frm,text='Login',command=lambda:validatehomescreen(frm,user_entry,pass_entry),width=10,font=('',12,''),bg='powder blue',bd=5)
    login_btn.place(x=530,y=300)
    
    reset_btn=Button(frm,text='Reset',command=lambda:resethome(user_entry,pass_entry),width=10,font=('',12,''),bg='powder blue',bd=5)
    reset_btn.place(x=650,y=300)
    
# RESETTING THE HOMEPAGE AND THE FEEDBACK PAGE 

def resethome(eu,ep):
    eu.delete(0,END)
    ep.delete(0,END)
    eu.focus()

def resetFeedback(et):
    et.delete(0,END)
    et.focus()
    
# VALIDATING THE HOME PAGE TO ALLOW ACCESS TO VERIFIED USERS
    
def validatehomescreen(frm,eu,ep):
    u=eu.get()
    p=ep.get()
    if(len(u)==0 or len(p)==0):
        messagebox.showwarning('Validaton failed',"username/password cant be empty")
        return
    else:
        if (u == 'admin' and  p== 'secret'):
            messagebox.showinfo('info','Correct Login')
            welcomeuser(frm,u)
        else:
            messagebox.showinfo('info','Invalid Login')
            
# CLEANING THE TEXT BY REMOVING PUNCTUATIONS AND CONVERTING THEM INTO TOKENS
    
def clean_text(msg):
    msg=re.sub(f"[{string.punctuation}]","",msg)
    msg=msg.lower()
    list_wrd=word_tokenize(msg)
    temp_list=[]
    for w in list_wrd:
        if(w not in words):
            temp_list.append(ps.stem(w))
    return " ".join(temp_list)

# TRAINING THE DATA 

df=pd.read_csv(r"C:\Users\A K SHARMA\Downloads\dataset\sentiment\Restaurant_Reviews.txt",delimiter='\t')
df.Review=df.Review.map(clean_text)
cv=CountVectorizer(binary=False,ngram_range=(1,2))
clf=MultinomialNB()
X=cv.fit_transform(df.Review).toarray()
y=df.Liked
clf.fit(X,y)            

def welcomeuser(prvfrm,u):
    prvfrm.destroy()
    frm=Frame(win,bg='yellow')
    frm.place(x=0,y=100,relwidth=1,relheight=1)
    
    singlefeedback_btn=Button(frm,text='Single Feedback',command=lambda:predict_single(frm),width=40,font=('',12,''),bg='powder blue',bd=5)
    singlefeedback_btn.place(x=550,y=190)

    bulkfeedback_btn=Button(frm,text='Bulk Feedback',command=lambda:predict_bulk(frm),width=40,font=('',12,''),bg='powder blue',bd=5)
    bulkfeedback_btn.place(x=550,y=280)
    
# ANALYZING THE SENTIMENTS OF A REVIEW GIVEN BY USER
        
def predict_single(prvfrm):
    frm=Frame(win,bg='yellow')
    prvfrm.destroy()
    global Entry1
    _bgcolor = '#00ff80'  # X11 color: 'gray85'
    _fgcolor = '#000000'  # X11 color: 'black'
    _compcolor = '#d9d9d9' # X11 color: 'gray85'
    _ana1color = '#d9d9d9' # X11 color: 'gray85'
    _ana2color = '#ececec' # Closest X11 color: 'gray92'
    font14 = "-family {Segoe UI} -size 14 -weight bold -slant roman -underline 0 -overstrike 0"
    font15 = "-family {Segoe UI} -size 12 -weight bold -slant roman -underline 0 -overstrike 0"
    font18 = "-family {Segoe UI} -size 28 -weight bold -slant roman -underline 1 -overstrike 0"
    win.geometry("856x450+209+115")
    win.title=Label(win,text='Restaurant Review Analysis',font=('',50,'bold'),fg='black',bg=bgclr)
    win.configure(background="#d9d9d9")
    Label1 = tk.Label()
    Label2 = tk.Label()
    Label1.place(relx=0.188, rely=0.244, height=31, width=185)
    Label1.configure(background="#d9d9d9")
    Label1.configure(font=font14)
    Label1.configure(text='''FEEDBACK:''',fg='blue')

    Entry1 = tk.Entry()
    Entry1.place(relx=0.327, rely=0.244,height=30, relwidth=0.276)
    Entry1.configure(background="white")
    Entry1.configure(width=264)
    Entry1.configure('',font=font14)
    
    Button1 = tk.Button(command=lambda:predict())
    Button1.place(relx=0.387, rely=0.4, height=35)
    Button1.configure(background="#d9d9d9")
    Button1.configure(font=font15)
    Button1.configure(text='''PREDICTION''')
    
    Button2 = tk.Button(command=lambda:resetFeedback(Entry1))
    Button2.place(relx=0.487, rely=0.4, height=35)
    Button2.configure(background="#d9d9d9")
    Button2.configure(font=font15)
    Button2.configure(text='''RESET''')

    def predict():
        if(len(Entry1.get())==0):
            messagebox.showwarning('Validaton failed',"Field cannot be empty")
        else:
            test=clean_text(Entry1.get())
            test_X=cv.transform([test])
            pred=clf.predict(test_X.toarray())
            if(pred[0]==0):
                Label2.configure(text="Did not Like")
            if(pred[0]==1):
                Label2.configure(text="Liked")
        return Label2       
    
 
    Label2.place(relx=0.388, rely=0.556, height=31, width=123)
    Label2.configure(background="#d9d9d9")
    Label2.configure(disabledforeground="#a3a3a3")
    Label2.configure(font=font14)
    Label2.configure(foreground="#000000")
    Label2.configure(text='''''',fg='blue')
    
# BROWSING AND LOADING A FILE CONSISTING OF BUNCH OF REVIEWS THEN ANALYZING SENTIMENTS OF EACH REVIEW AND DOWNLOADING THE FINAL ON SPECIFIED LOCATION

def predict_bulk(prvfrm):
    frm=Frame(win,bg='yellow')
    prvfrm.destroy()
    global Entry1
    _bgcolor = '#00ff80'  # X11 color: 'gray85'
    _fgcolor = '#000000'  # X11 color: 'black'
    _compcolor = '#d9d9d9' # X11 color: 'gray85'
    _ana1color = '#d9d9d9' # X11 color: 'gray85'
    _ana2color = '#ececec' # Closest X11 color: 'gray92'
    font14 = "-family {Segoe UI} -size 14 -weight bold -slant roman -underline 0 -overstrike 0"
    font15 = "-family {Segoe UI} -size 12 -weight bold -slant roman -underline 0 -overstrike 0"
    font18 = "-family {Segoe UI} -size 28 -weight bold -slant roman -underline 1 -overstrike 0"
    win.geometry("856x450+209+115")
    win.title=Label(win,text='Restaurant Review Analysis',font=('',50,'bold'),fg='black',bg=bgclr)
    win.configure(background="#d9d9d9")
    
    def UploadAction(event=None):
        filename = filedialog.askopenfilename()
        df1=pd.read_csv(filename,delimiter='\t')
        Entry1.insert(tk.END, filename)
        result = map(lambda predict:df1)
        #df1.Review=df1.Review.map(clean_text)
        #cv1=CountVectorizer(binary=False,ngram_range=(1,2))
        #clf1=MultinomialNB()
        #X1=cv.fit_transform(df.Review).toarray()
        #y1=df.Liked
        #clf1.fit(X1,y1)
        
        
    Label1 = tk.Label()
    Label2 = tk.Label()
    Label1.place(relx=0.188, rely=0.244, height=31, width=185)
    Label1.configure(background="#d9d9d9")
    Label1.configure(font=font14)
    Label1.configure(text='''FEEDBACK:''',fg='blue')
    
    Entry1 = tk.Entry()
    Entry1.place(relx=0.307, rely=0.244,height=30, relwidth=0.180)
    Entry1.configure(background="white")
    Entry1.configure(width=264)
    Entry1.configure('',font=font14)
    
    Button1 = tk.Button(command=lambda:UploadAction())
    Button1.place(relx=0.480, rely=0.244,height=30, relwidth=0.110)
    Button1.configure(background="#d9d9d9")
    Button1.configure(font=font15)
    Button1.configure(text='''Open''')
    
    Button2 = tk.Button(command=lambda:predict())
    Button2.place(relx=0.387, rely=0.4, height=35)
    Button2.configure(background="#d9d9d9")
    Button2.configure(font=font15)
    Button2.configure(text='''Prediction''')
    
    Button3 = tk.Button(command=lambda:resetFeedback(Entry1))
    Button3.place(relx=0.487, rely=0.4,height=35)
    Button3.configure(background="#d9d9d9")
    Button3.configure(font=font15)
    Button3.configure(text='''Reset''')
    
    
    def predict(a):
        if(len(a)==0):
            messagebox.showwarning('Validaton failed',"Field cannot be left empty")
        else:
            test=clean_text(a.get())
            test_X=cv.transform([test])
            pred=clf.predict(test_X.toarray())
            print(test_X)
            if(pred[0]==0):
                Label2.configure(text="Did not Like")
            if(pred[0]==1):
                Label2.configure(text="Liked")
        return Label2
    
    
    Label2.place(relx=0.388, rely=0.556, height=31, width=123)
    Label2.configure(background="#d9d9d9")
    Label2.configure(disabledforeground="#a3a3a3")
    Label2.configure(font=font14)
    Label2.configure(foreground="#000000")
    Label2.configure(text='''''',fg='blue')


homescreen()
win.mainloop()
```
